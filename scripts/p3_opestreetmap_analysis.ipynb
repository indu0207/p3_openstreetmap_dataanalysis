```python
import xml.etree.cElementTree as ET
import re

osm_file = open("/Users/indur/Downloads/new-delhi_india.osm", "r")

city_names = set()

streets = set()
street_mapping = { 
            "Ave": "Avenue",
            "Ave.": "Avenue",
            "Rd": "Road",
            "Rd.": "Road",
            "St": "Street",
            "St.": "Street"
            }
street_type_re = re.compile(r'\b\S+\.?$', re.IGNORECASE)

lower = re.compile(r'^([a-z]|_)*$')
lower_colon = re.compile(r'^([a-z]|_)*:([a-z]|_)*$')
problemchars = re.compile(r'[=\+/&<>;\'"\?%#$@\,\. \t\r\n]')

problem_keys = []
keys = {"lower": 0, "lower_colon": 0, "problemchars": 0, "other": 0}

for event, elem in ET.iterparse(osm_file, events=("start",)):

    #city names audit
    if elem.tag == "node" :
        for tag in elem.iter("tag"):
            if tag.attrib["k"] == "addr:city":
                city_names.add(tag.attrib["v"])
                
    if elem.tag == "node" or elem.tag == "way":
        for tag in elem.iter("tag"):
            if tag.attrib["k"] == "addr:street":
                streets.add(tag.attrib["v"])

    if elem.tag == "tag":
        v = elem.attrib['k']
        m = lower.search(v) 
        if m:
            keys["lower"] += 1
        else:
            m = lower_colon.search(v)
            if m:
                keys["lower_colon"] += 1
            else:
                m = problemchars.search(v)
                if m:
                    keys["problemchars"] += 1
                    problem_keys.append(elem.attrib)
                else:
                    keys["other"] += 1
                
#problem characters
print problem_keys
#general stats
print  keys

exit;

#Looking at the set, there are some typos, street and colony names in the city field. This needs a clean up.
#final list what every Address needs to be replaced.
#Also, some of errors needs to be manually fixed
#1. Spelling mistakes
#2. Cases where the state is also mentioned along with the city names

for each_city in city_names:
    temp_arr = each_city.split(', ')
    print each_city + " => " + temp_arr[len(temp_arr) - 1] 



#corrections needed on the street names
street_mapping_keys = street_mapping.keys()   
for each_street in streets:
    m = street_type_re.search(each_street)
    if m:
        temp_st_words = [m.group()]
        for every_st_word in temp_st_words:
            if every_st_word in street_mapping_keys:
                print each_street 

          
```
